```python
import pandas as pd
import numpy as np
```

```python
df_train=pd.read_csv('train-g.csv')
df_test=pd.read_csv('test_g.csv')
fulfilment=pd.read_csv('fulfilment_center_info.csv')
meal_info=pd.read_csv('meal_info.csv')
```

```python
df_final1=pd.merge(df_full, fulfilment[['center_id','center_type', 'op_area']], how='outer', on='center_id')
```

```python
df=pd.merge(df_final1, meal_info,how='outer', on='meal_id')
```

```python
df.corr()
```

```python
df.category.value_counts()
```

```python
df.groupby(['center_type','cuisine'])['num_orders'].mean().plot('bar')
```

```python
for x in df.columns:
    if x[-1]=='d' and x[-2]=='i':
        df.drop([x],1,inplace=True)
            
```

```python
df.columns
```

```python
df=pd.get_dummies(df, columns=['center_type', 'cuisine'], dummy_na=False)
```

```python
df.drop('category', 1, inplace=True)
```

```python
df.describe()
```

```python
df.groupby('emailer_for_promotion')['num_orders'].mean()
```

```python
df.groupby('homepage_featured')['num_orders'].mean()
```

```python
df_train=df[df.num_orders.isnull()==False]
```

```python
df_test=df[df.num_orders.isnull()==True]
```
